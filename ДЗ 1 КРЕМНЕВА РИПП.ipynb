```python
import pandas as pd   
import numpy as np     
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
%matplotlib inline
```

```python
#ЗАГРУЖАЕМ ФАЙЛЫ
ratings=pd.read_csv("ratings.csv") 
movies=pd.read_csv("movies.csv") 
tags=pd.read_csv("tags.csv")
```

```python
#ОБЪЕДИНЯЕМ ДВА ДАТАФРЕЙМА
movies = movies.merge(ratings, on = 'movieId', how = 'inner')
movies.head()
```

```python
#СОЗДАЕМ ДАТАСЕТ С ФИЛЬМАМИ,ЮЗЕРАМИ И ИХ ОЦЕНКАМИ
movies = movies[['userId','movieId','rating']]
movies.head()
```

```python
conda install -c conda-forge scikit-surprise
```

```python
from surprise import SVD
from surprise import Dataset, Reader
from surprise.model_selection import cross_validate
```

```python
reader=Reader(rating_scale=(1,5))
data = Dataset.load_from_df(movies, reader)
```

```python
algo = SVD()
cross_validate(algo, data, measures=['MSE'], cv=8, verbose=True)
```

Итак, MSE = 0.7559 - это лучшее значение, которое нам удалось получить
